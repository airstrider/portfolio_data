## Part 5: Evaluating the models.

This `Part 5` will review and evaluate the 4 models.
1. Loading the result data made from the previous parts. 
2. Visualizing the result data with `Seaborn`.

### 1. Loading the result data

#### 1.1. Import packages

Import relevant Python packages.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 1.2. Load the dataset

`Pandas` is used to load the `ml_results.csv`. The resulting pandas `DataFrame` is saved in a variable named `results`.

```python
results = pd.read_csv('ml_results.csv').drop(columns=['Unnamed: 0'])
results
```

### 2. Visualizing the result data

#### 2.1. Create bar charts

Use `subplots()` function to organize the 4 charts in 1 frame.

```python
fig, axs = plt.subplots(2,2,figsize=(10,5))
axs[0][0].set_title('metric: Accuracy')
axs[0][1].set_title('metric: Precision')
axs[1][0].set_title('metric: Recall')
axs[1][1].set_title('metric: F1')
sns.barplot(ax=axs[0][0], data=results, x='Accuracy', y='Model', width=0.5, ).set(xlim=(.93,.97))
sns.barplot(ax=axs[0][1], data=results, x='Precision',y='Model', width=0.5).set(xlim=(.93,.97))
sns.barplot(ax=axs[1][0], data=results, x='Recall',   y='Model', width=0.5).set(xlim=(.93,.97))
sns.barplot(ax=axs[1][1], data=results, x='F1',       y='Model', width=0.5).set(xlim=(.93,.97))
plt.tight_layout();
```

### 3. Evaluating the models based on the metrics.

#### 3.1. Review the bar charts

* All the metrics show `Tuned XGBoost` model performed much better than others.
* For `Precision`, `Tuned Decision Tree` model performed slightly better than `Tuned Random Forest`.
* For `Recall`, `Tuned Decision Tree` model is the worst one, even `Decision Tree`.
* When it comes to identifying claims, `Recall` score tends to reflect the real situations bettern than others.


From the all observations, we can conclude `Tuned XGBoost` model is the champion.
